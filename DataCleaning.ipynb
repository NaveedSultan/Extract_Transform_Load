```python
# handling Missing values
df.dropna(subset=['Name'],inplace=True)
```

```python
df['Age'].unique()
```

```python
# cleaning the age column
import re

def extract_age(age):
    age_num=re.findall('[0-9]+',str(age))
    if len(age_num)>0:
        return age_num[0]
    else:
        return age


df['Age']=df['Age'].apply(lambda x:extract_age(x))
```

```python
# we found Nan values in age column
df['Age'].head(50)
```

```python
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
```

```python
df['Age'].unique()
```

```python
age_median=df['Age'].median()
df['Age'].fillna(age_median,inplace=True)
```

```python
df['Age'].unique()
```

```python
# coz of NaN values age column was in float so we converted into int dtype
df['Age']=df['Age'].astype(int)
```

```python
df['Age']
```

```python
# Age is converted into int dtype
df.info()
```

```python
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(),inplace=True)
```

```python
df['Feedback_Score'].fillna(df['Feedback_Score'].mode()[0],inplace=True)
```

```python
# Filling most frequent data in City & Gender

for col in ['City','Gender']:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
# dropping null values in email column
df.dropna(subset=['Email'],inplace=True)
```

```python
# filling null values in Last_Purchase_Date
df['Last_Purchase_Date'].ffill(inplace=True)
```

```python
# Handled the null values 
df.isnull().sum()
```

```python
df['Gender']=df['Gender'].str.lower().str.strip()
df['Gender']
```

```python
df['Gender'].replace({'femaleale':'female','maleale':'male','f':'female','m':'male'},inplace=True)
```

```python
df['City']=df['City'].str.lower()
```

```python
df.head(50)
```

```python
df['Customer_ID'].value_counts()
```

```python
# handling Duplicated rows and columns
df[df.duplicated()].shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[df.duplicated()].shape
```

```python
df['Customer_ID'].value_counts()
```

```python
# due to inconsistent data there was multiple same Customer_ID
df.drop_duplicates(subset=['Customer_ID'],keep='first',inplace=True)
```

```python
# Correcting the Datatype
df['Age']
```

```python
df.info()
```

```python
# converted into datetime type
df['Signup_Date']=pd.to_datetime(df['Signup_Date'])
df['Last_Purchase_Date']=pd.to_datetime(df['Last_Purchase_Date'])
```

```python
# Handling Outliers
Q1=df['Purchase_Amount'].quantile(0.25)
Q3=df['Purchase_Amount'].quantile(0.75)
IQR=Q3-Q1

outliers=df[(df['Purchase_Amount']<Q1-1.5*IQR)|(df['Purchase_Amount']>Q3+1.5*IQR)]

print(outliers.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
cols=['Purchase_Amount','Feedback_Score','Age']
colors=['#3498db','#2ecc71','#e74c3c']

for ax,col,color in zip(axes,cols,colors):
    ax.boxplot(df[col].dropna(),patch_artist=True,boxprops=dict(facecolor=color,alpha=0.6))
    ax.set_title(f'{col}',fontsize=13,fontweight='bold')
    ax.set_ylabel('Value')
plt.suptitle('Outliers-Dectection',fontweight='bold')
plt.tight_layout()
plt.show()
```
